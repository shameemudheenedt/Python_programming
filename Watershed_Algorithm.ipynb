Watershed Algorithm Implimentation

```python
#Import packages
from skimage import io
from skimage import data
```

```python
#Load coin dataset
coin = data.coins()
```

```python
#Sobel mask for edge Detection
from skimage.filters import sobel
elevation_map = sobel(coin)
```

```python
#Define Markers
import numpy as np
import matplotlib.pyplot as plt
markers=np.zeros_like(coin)
markers[coin<30]=1
markers[coin>150]=2
plt.imshow(markers)
plt.show()
```

```python
#Import watershed package
from skimage.segmentation import watershed
se=watershed(elevation_map,markers)
plt.imshow(se,cmap='gray')
```

```python
from scipy import ndimage as ndi
from skimage.color import label2rgb
seg_c = ndi.binary_fill_holes(se-1)
labeled,_ = ndi.label(seg_c)
im_ov = label2rgb(labeled,image=coin)
plt.imshow(im_ov)
plt.show()
```
