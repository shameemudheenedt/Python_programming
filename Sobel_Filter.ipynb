```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image
image = Image.open('C:/Users/Computer Science Lap/Downloads/MICC-F8multi/1 (5).jpg')
plt.imshow(image)
```

```python
image_arr = np.asarray(image) # convert image to numpy array
image_arr.shape
gray = cv2.cvtColor(image_arr, cv2.COLOR_BGR2GRAY)
Horizontal_kernel = np.array([[-1,-1,-1],[2,2,2],[-1,-1,-1]])
Vertical_kernel = np.array([[-1,2,-1],[-1,2,-1],[-1,2,-1]])
blurimg = cv2.filter2D(gray,-1,Horizontal_kernel)
blurimg1 = cv2.filter2D(gray,-1,Vertical_kernel)
plt.imshow(blurimg, cmap='gray')
plt.title('Horizontal Filter Applied image')
```

```python
plt.imshow(blurimg1, cmap='gray')
plt.title("Vertical mask applied image")
```

```python
sobelx=cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=5)
sobely=cv2.Sobel(gray,cv2.CV_64F,0,1,ksize=5)
mag,direction = cv2.cartToPolar(sobelx,sobely,angleInDegrees =True)
sobelG = np.hypot(sobelx,sobely)
```

```python
plt.imshow(sobelx,cmap='gray')
plt.title("Sobel mask applied image")
plt.show()
```
